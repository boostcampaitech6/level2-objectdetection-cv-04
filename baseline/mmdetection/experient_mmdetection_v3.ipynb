```python
# Copyright (c) OpenMMLab. All rights reserved.
import argparse
import os
import os.path as osp
import mmengine
from mmengine.config import Config, DictAction
from mmengine.registry import RUNNERS
from mmengine.runner import Runner

from mmdet.utils import setup_cache_size_limit_of_dynamo
```

```python
# config file 들고오기
cfg = Config.fromfile("./configs/Outliers/faster-rcnn_r50_fpn_1x_trash.py")

root='../../dataset/'
print(cfg.keys())
cfg.train_dataloader.dataset.data_prefix = dict(img="")
cfg.train_dataloader.dataset.ann_file = "k-fold/train_fold_7.json"

cfg.train_dataloader.batch_size = 4

cfg.seed = 2024
cfg.gpu_ids = [0]
cfg.work_dir = './work_dirs/faster_rcnn_r50_fpn_1x_trash'

cfg.model.roi_head.bbox_head.num_classes = 10

cfg.optimizer_config = dict(grad_clip = dict(max_norm=35, norm_type=2))
cfg.checkpoint_config = dict(max_keep_ckpts=3, interval=1)
print(cfg.custom_hooks)
```

```python
# build the runner from config
if 'runner_type' not in cfg:
    # build the default runner
    runner = Runner.from_cfg(cfg)
else:
    # build customized runner from the registry
    # if 'runner_type' is set in the cfg
    runner = RUNNERS.build(cfg)

# start training
runner.train()
```
